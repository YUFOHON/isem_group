```python
# Step 1
# Assume LTV is 90%

trans_price=int(input("Please input transaction price: ")) # Input transaction price
valu_price=int(input("Please input valuation price: "))  # Input valuation price
years=int(input("Please input Loan Tenor years: "))  # Input years


# calculate the down payment
def downpayment_cal(trans_price, valu_price):
    return 0.1*min(trans_price, valu_price)+ max(trans_price-valu_price, 0)

def agency_cal(trans_price):
    return 0.01*trans_price

# calculate the legal fee, according to table 1
def legalfee_cal(trans_price):
    if trans_price < 3000000:
        return 10000
    elif trans_price <= 5000000:
        return 11000
    elif trans_price < 10000000:
        return 12000
    else:
        return 0.001 * trans_price + 3000

    
# calculate the stamp duty fee, according to table 2
def stamp_cal(trans_price):
    if trans_price < 3000000:
        return 100
    elif trans_price < 3528240:
        return 100 + (trans_price - 3000000) * 0.1
    elif trans_price < 4500000:
        return 0.015 * trans_price
    elif trans_price < 4935480:
        return 67500 + (trans_price - 4500000) * 0.1
    elif trans_price < 6000000:
        return 0.0225 * trans_price
    elif trans_price < 6642860:
        return 135000 + (trans_price - 6000000) * 0.1
    elif trans_price < 9000000:
        return 0.03 * trans_price
    elif trans_price < 10080000:
        return 270000 + (trans_price - 9000000) * 0.1
    elif trans_price < 20000000:
        return 0.0375 * trans_price
    elif trans_price < 21739120:
        return 750000 + (trans_price - 20000000) * 0.1
    else:
        return 0.0425 * trans_price

# calculate the mortgage insurance, according to table 3,4
def mortgage_cal(trans_price, valu_price, years):
    
    if valu_price <= 6000000:  # when the property value up to HK$6 million
        rates = [[10, 1.25], [15, 1.48], [20, 1.79], [25, 2.03], [30, 2.16]]
        for i in rates:
            if i[0] == years:
                return min(trans_price, valu_price) * 0.9 * i[1] / 100
    else:  # when the property value up to HK$15 million
        rates = ([10, 1.46], [15, 1.72], [20, 2.08], [25, 2.35], [30, 2.50])
        for i in rates:
            if i[0] == years:
                return min(trans_price, valu_price) * 0.9 * i[1] / 100


print()  # print a blank row

down_payment = downpayment_cal(trans_price, valu_price)
agency_fee = agency_cal(trans_price)
legal_fee = legalfee_cal(trans_price)
stamp_duty = stamp_cal(trans_price)
total=down_payment+agency_fee+legal_fee+stamp_duty
mortgage_insurance = mortgage_cal(trans_price, valu_price, years)

print("Downpayment =", downpayment_cal(trans_price, valu_price))
print("Agency fee =", agency_cal(trans_price))
print("Legal fee =", legalfee_cal(trans_price))
print("Stamp Duty =", stamp_cal(trans_price))
print("Total exoense =", total)
print("Mortgage Insurance =", mortgage_cal(trans_price, valu_price, years))

```

```python
# Step 2

def getTotal():
    # Ask to input required information
    trans_price=int(input("Please input transaction price: ")) # Input transaction price
    valu_price=int(input("Please input valuation price: "))  # Input valuation price
    years=int(input("Please input Loan Tenor years: "))  # Input years
 
    # Do the calculation and return total
    down_payment = downpayment_cal(trans_price, valu_price)
    agency_fee = agency_cal(trans_price)
    legal_fee = legalfee_cal(trans_price)
    stamp_duty = stamp_cal(trans_price)
    total=down_payment+agency_fee+legal_fee+stamp_duty
    mortgage_insurance = mortgage_cal(trans_price, valu_price, years)

    return total

total=getTotal()
initial_saving = int(input("Please enter your initial saving: "))
salary = int(input("Please enter your salary: "))
percentage_saved = int(input("Please enter your percentage saved: "))
r_rate = int(input("Please enter your annual return rate: "))

months = 0  # set a new variable to record requested months
cumulative_saving = initial_saving  # initialize the saving

# use 'while' loop to calculate how many months needed for the total expense required
while cumulative_saving < total:  # when cumulative saving larger than the total expense, it will terminate
    cumulative_saving += salary * percentage_saved / 100 \
                         + cumulative_saving * r_rate / 100 / 12
    months += 1
    
print("You should pay for loan in", months, "months")  # print the result
```

```python
from amortization.amount import calculate_amortization_amount
import datetime


# User inputs
transaction_cost = trans_price
valuation_price = valu_price
downpayment_percentage = 0.1  # Downpayment as a percentage of property cost (90% LTV)
buyer_salary_monthly = salary  # Buyer's monthly salary in HKD
mortgage_rebate_percentage = 0.03  # Bank rebate percentage
mortgage_annual_interest_compound_monthly = 0.1  # Mortgage annual  interest rate
mortgage_insurance_rate = 0.0025  # Mortgage insurance rate
downpayment = 0.1 * min(transaction_cost, valuation_price) + max(transaction_cost - valuation_price, 0)
loan_amount = valuation_price - downpayment



def calculate_dti_ratio(monthly_payment, buyer_salary_monthly):
    """
    Calculate the Debt-to-Income (DTI) ratio.
    """
    return monthly_payment / buyer_salary_monthly

def calculate_stress_test_ratio(monthly_payment, buyer_salary_monthly):
    """
    Calculate the Stress Test ratio.
    """
    stress_test_payment = monthly_payment * (1 + 0.02)

    return stress_test_payment / buyer_salary_monthly

def assess_dti_ratio(loan_tenor, monthly_payment, buyer_salary_monthly):
    """
    Assess the DTI Ratio for a given loan tenor.
    """
    dti_ratio = calculate_dti_ratio(monthly_payment, buyer_salary_monthly)
    return dti_ratio <= 0.5

def assess_stress_test_ratio(loan_tenor, monthly_payment, buyer_salary_monthly):
    
    # Assess the Stress Test ratio for a given loan tenor.
    
    stress_test_ratio = calculate_stress_test_ratio(monthly_payment, buyer_salary_monthly)
    return stress_test_ratio <= 0.6

def calculate_minimum_salary(loan_amount, loan_tenor, mortgage_annual_interest_compound_monthly):
    """
    Calculate the minimum required salary to fulfill the DTI ratio and Stress Test requirements.
    """
    monthly_payment = calculate_amortization_amount(
        loan_amount,
        mortgage_annual_interest_compound_monthly,
        loan_tenor * 12
    )

    min_salary_monthly = monthly_payment / 0.5  # Minimum salary to meet 50% DTI ratio
    stress_test_payment = monthly_payment * (1 + 0.02)
    min_salary_monthly_stress_test = stress_test_payment / 0.6  # Minimum salary to meet 60% Stress Test ratio

    return min_salary_monthly, min_salary_monthly_stress_test

loan_tenors = [10, 15, 20, 25, 30]  # Loan tenor options in years
# Mortgage insurance rates based on loan tenor
mortgage_insurance_rates_below_six_million = {
    10: 1.25,
    15: 1.48,
    20: 1.79,
    25: 2.03,
    30: 2.16
}
mortgage_insurance_rates_above_six_million = {
    10: 1.46,
    15: 1.72,
    20: 2.08,
    25: 2.35,
    30: 2.50
}

def step3():
    #buyer_salary_monthly = float(input("Please enter the buyer's monthly salary in HKD: "))
    print("Buyer's monthly salary: ", buyer_salary_monthly)
    #ask user to input the transaction cost
    #transaction_cost = float(input("Please enter the transaction cost in HKD: "))
    print("Transaction cost: ", transaction_cost)
    #ask user to input the valuation price
    #valuation_price = float(input("Please enter the valuation price in HKD: "))
    print("valuation cost: ", valuation_price)
    mortgage_annual_interest_compound_monthly = float(input("Please enter the Mortgage annual interest compound monthly: "))
    print("Mortgage annual interest compound monthly: ", mortgage_annual_interest_compound_monthly)
    # ask user to input the mortgage rebate percentage
    mortgage_rebate_percentage = float(input("Please enter the mortgage rebate percentage: "))
    print("Mortgage rebate percentage: ", mortgage_rebate_percentage)
    
    #Get the current timestamp
    current_timestamp = datetime.datetime.now()
    # Print the timestamp
    print(f"The current timestamp is: {current_timestamp}")
    print("Loan Tenor | Monthly Payment | Loan Amount     | Minimum Salary Pass DTI | Minimum Salary Pass Stress Test | DTI Ratio Met | Stress Test Met")
    print("-" * 140)

    for loan_tenor in loan_tenors:

        # Calculate monthly payment
        monthly_payment = calculate_amortization_amount(
            loan_amount,
            mortgage_annual_interest_compound_monthly,
            loan_tenor * 12
        )

        # Assess DTI Ratio
        dti_ratio_met = assess_dti_ratio(loan_tenor, monthly_payment, buyer_salary_monthly)
        # Assess Stress Test
        stress_test_ratio_met = assess_stress_test_ratio(loan_tenor, monthly_payment, buyer_salary_monthly)

        # Calculate loan amount based on 90% LTV
        loan_amount_base = 0.9 * min(transaction_cost, valuation_price)
        # Calculate mortgage insurance amount for loan amount above 70% LTV
        mortgage_insurance_rates = mortgage_insurance_rates_above_six_million if min(transaction_cost, valuation_price) > 6000000 else mortgage_insurance_rates_below_six_million
        mortgage_insurance_rate = mortgage_insurance_rates[loan_tenor]
        mortgage_insurance_amount = (min(transaction_cost, valuation_price) * 0.9 * mortgage_insurance_rate)/100
        # Calculate final loan amount
        current_loan_amount = loan_amount_base + mortgage_insurance_amount - (valuation_price * mortgage_rebate_percentage)
        # Calculate the minimum required salary
        min_salary_monthly = calculate_minimum_salary(
            loan_amount,
            loan_tenor,
            mortgage_annual_interest_compound_monthly
        )
        print(f"{loan_tenor:>5}-year | {monthly_payment:>10,.2f} HKD | {current_loan_amount:>10,.2f} HKD | {min_salary_monthly[0]:>18,.2f} HKD | {min_salary_monthly[1]:>26,.2f} HKD | {dti_ratio_met!s:>13} | {stress_test_ratio_met!s:>13}")
        print("-" * 140)
```
